```python
# Checking structures of the h5 file
with h5py.File('../Data/Filtered_feature_bc_matrix.h5', 'r') as file:
    print("Keys: %s" % list(file.keys()))
    a_group_key = list(file.keys())[0]

    matrix_group = file[a_group_key]

    features_group = matrix_group['features']
    print("Features Group Keys: %s" % list(features_group.keys()))

    for feature_key in features_group.keys():
        print(f"Feature Key: {feature_key}")
        feature_item = features_group[feature_key]
        
        if isinstance(feature_item, h5py.Dataset):
            print(feature_item[:])
        else:
            print("Subgroup within features:", feature_item)
```

```python
# Using hdf5 file to find specific gene's expression value
with h5py.File('../Data/Filtered_feature_bc_matrix.h5', 'r') as file:
    matrix_group = file['matrix']
    
    # Read relevant datasets
    barcodes = matrix_group['barcodes'][:]
    data = matrix_group['data'][:]
    indices = matrix_group['indices'][:]
    indptr = matrix_group['indptr'][:]
    
    # Access the 'features' subgroup to get feature names and other attributes
    features_group = matrix_group['features']
    feature_names = features_group['name'][:]

    # Print the first 5 pieces of linked data
    for i in range(1):
        # Find the feature index range for each barcode
        start = indptr[i]
        end = indptr[i + 1]
        print(f"Barcode: {barcodes[i].decode()}")
        
        # Iterate over each feature in the range
        for j in range(start, end):
            feature_index = indices[j]
            expression_value = data[j]
            feature_name = feature_names[feature_index].decode()
            print(f"  Feature: {feature_name}, Expression Value: {expression_value}")
```
